```python
import os
from datetime import datetime

import matplotlib.pyplot as plt

from skimage.io import imread, imshow
```

## Starts, or restarts Raspberry Pi camera preview window

```python

from picamera2 import Picamera2
try:
    cam
except NameError:
    pass
else:
    cam.close()
cam = Picamera2()

cam.start(show_preview=True)
# uncomment following lines to disable auto white balance and auto exposure
# and manually set camera parameters
cam.set_controls({'AwbEnable':True,'AeEnable':False})
cam.set_controls({'ExposureTime':10000, 'AnalogueGain':3.0})
```

```python
# capture 1 frame to numpy array
img = cam.capture_array("main")
# display image
img_plot = plt.imshow(img)
```

## Capture 1 frame and save as file

```python

# Get the current timestamp
now = datetime.now()
# Examples of different formats
formatted_date = now.strftime('%Y%m%d_%H%M%S')
filename = os.path.expanduser(f'~/Pictures/{formatted_date}.tif')
metadata = cam.capture_file(filename)
```

## Read image with scikit-image and display it _ ready to analyse!!

```python
images = imread(filename)
imshow(images)
print(images.shape)
```

```python
# close camera and preview window
cam.close()
```
