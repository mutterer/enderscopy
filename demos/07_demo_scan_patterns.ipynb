```python
from enderscope import ScanPatterns

origin = [25, 100] # coordinates of first position
steps = [38, -25] # x and y shifts to next item
positions = ScanPatterns.raster(4, 3) * steps + origin
ScanPatterns.plot_path(positions, field=(40, 30),title='raster scan')
```

```python
positions = ScanPatterns.snake(4, 4) * [33.33, -25] + [50, 100]
ScanPatterns.plot_path(positions, title='snake scan')
```

```python
positions = ScanPatterns.random(20)
ScanPatterns.plot_path(positions, title='random scan')
```

```python
positions = ScanPatterns.spiral(16) * [30, 20]  + [100, 75]
ScanPatterns.plot_path(positions,field=(34, 24),title='spiral scan')
```
